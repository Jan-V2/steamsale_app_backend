```python
from dataclasses import dataclass
import json

@dataclass
class Row:
    title: str
    app_id:int
    price_old:float
    price_new:float
    discount_percent:int
    reviews_count:int
    reviews_percent_good:int
    is_bundle:bool

data = open("./downloaded/EU.json").read()
data = json.loads(data)
data = data["items"]
data 
```

```python
rows = []
for item in data:
    rows.append(Row(
        title=item["titles"],
        app_id=item["appids"], #there can be multible apparrentlly
        price_old=item["old_price"],
        price_new=item["new_price"],
        discount_percent=item["discount_percents"],
        reviews_count=item["n_user_reviews"],
        reviews_percent_good=item["percent_reviews_positive"],
        is_bundle=item["is_bundle"]
    ))
rows[0]
```

```python
def print_n(array, n):
    for i in range(n):
        print(f"old price:{array[i].discount_percent}, new price:{array[i].price_new} percent:{array[i].discount_percent} title: {array[i].title}")
        
```

```python
min_discount = 60
min_reviews = 1000
max_new_price = 30.0
min_new_price = 10.0


filtered = filter(lambda i:i.discount_percent > min_discount,rows)
filtered = filter(lambda i:i.reviews_count > min_reviews, filtered)
filtered = filter(lambda i:i.price_new < max_new_price, filtered)
filtered = filter(lambda i:i.price_new > min_new_price, filtered)

sorted_list = list(filtered)
sorted_list.sort(key=lambda i:i.discount_percent, reverse=True)

print_n(sorted_list, 100)
```
